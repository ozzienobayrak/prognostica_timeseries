```python
import pandas as pd
import psycopg2
import os
from dotenv import load_dotenv
```

```python
#!pip install SQLAlchemy==2.0.15
from sqlalchemy import create_engine

#read the database string from the .env
load_dotenv()

DB_STRING = os.getenv('DB_STRING')

db = create_engine(DB_STRING)
```

```python
from sqlalchemy import text
import pandas as pd

query_string = """
SELECT 
    date_trunc('month', to_date(date, 'DD.MM.YYYY'))::date AS month,
    EXTRACT(YEAR FROM to_date(date, 'DD.MM.YYYY')) AS year,
    SUM(value) AS monthly_revenue
FROM takemehome.prognostica_daten_kunden
WHERE customer NOT LIKE '%DEK'
GROUP BY 1, 2
ORDER BY 1, 2;
"""

with db.connect() as conn:
    df_sqlalchemy = pd.read_sql_query(text(query_string), conn)

df_sqlalchemy.head()
```

```python
df_sqlalchemy.info()
```

```python
#export the data to a csv-file
df_sqlalchemy.to_csv('data/prog_customer.csv',index=False)
```

```python
#import the data from a csv-file
df_customer = pd.read_csv('data/prog_customer.csv')
```
